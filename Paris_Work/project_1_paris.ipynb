# Cleaning the Data

```python
# Channel Natasha Bedingfield and release your inhibitions... and drop the '% Other' column
df = df.drop(columns = "% Undeclared")
# Remove rows with Company name as "Apple (excluding undeclared)" as this is a duplicate of "Apple"
df = df[df.Company != "Apple (excluding undeclared)"]
df
```

```python
# Sum the nulls in each column
print(df.isnull().sum())
```

```python
df.loc[3, '% Other'] = '0'
df
```

```python
print(df.isnull().sum())
```

```python
# Replace all null values with 0 in '% Other' column
df['% Other'] = df['% Other'].fillna(0)

# Replace all '<1' values with 0 in '% Other' columns
df['% Other'] = df['% Other'].replace('<1', 0)

# Replace all '-' values with 0 in '% Other' column
df['% Other'] = df['% Other'].replace('-', 0)

# Replace all the '-' calues with 0 in '% Asian' column
df['% Asian'] = df['% Asian'].replace('-', 0)

# Replace all the '-' calues with 0 in '% Latino' column
df['% Latino'] = df['% Latino'].replace('-', 0)

# Replace all the '-' calues with 0 in '% Black' column
df['% Black'] = df['% Black'].replace('-', 0)

# Replace all the '-' calues with 0 in '% Multi' column
df['% Multi'] = df['% Multi'].replace('-', 0)
```

```python
# Check to see if replacement code was successful
# print(df.isnull().sum())
df.info()
```

```python
# Convert to floats
df['Female %'] = df['Female %'].astype(float)
df['Male %'] = df['Male %'].astype(float)
df['% White'] = df['% White'].astype(float)
df['% Asian'] = df['% Asian'].astype(float)
df['% Latino'] = df['% Latino'].astype(float)
df['% Black'] = df['% Black'].astype(float)
df['% Multi'] = df['% Multi'].astype(float)
df['% Other'] = df['% Other'].astype(float)
```

```python
df.to_csv("DiversityinTechCompanies_PL.csv")
```

```python
# Import dependencies
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
from sklearn import datasets
import pandas as pd

print(plt.style.available)
```

```python
# Plot women hired vs. latino people hired
x_values = df['Female %']
y_values = df['% Latino']

# Style
plt.style.use('bmh')
# Add linear regression equation and line to plot (with code from regression_solution activity)
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values, y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq, xy=(20, 15), fontsize=15, color="red") # line of code from chat gpt with help from Christopher Madden (TA)
plt.xlabel('Women Hired (%)')
plt.ylabel('Latino People Hired (%)')
print(f"The r-value is: {rvalue}")
plt.show()
```
